```python
import numpy as np
x = np.linspace(0.5, 11.5, 12)
y = [2,5,7,7,2,6,6,2,7,3,6,8]

plt.bar(x, y, label="bar 1")
plt.title("bar chart")
plt.xlabel("month")
plt.ylabel("income")
plt.ylim(0, 10)
plt.xlim(0.5, 12.5)
plt.legend()
```

```python
x1 = np.linspace(1, 9, 5)
y1 = [2, 6, 6, 4, 2]

x2 = np.linspace(2, 10, 5)
y2 = [4, 6, 3, 7, 5]

# use single character, eg. 'r', 'g', or word like 'blue' or hexical number like '#32BEBE'
# color k for black, 'c' for cyan
plt.bar(x1, y1, label="bar 1", color="k")
plt.bar(x2, y2, label="bar 2", color="c")
plt.legend()
```

```python
import random
mu, sigma = 50, 20
mean = mu
std = sigma
x = np.random.normal(loc=mean, scale=std, size=2000)
x = [random.gauss(50, 20) for i in range(2000)]
bins = [i for i in range(0, 110, 10)]
plt.hist(x, bins, histtype='bar', rwidth=0.7)
```
